```python
import pandas as pd

f1_df = pd.read_csv(
    "../data/processed/f1_analytics_dataset.csv"
)

print(f1_df.shape)
```

```python
f1_df["valid_grid"] = (
    f1_df["grid"] > 0
)
```

```python
f1_df["position_gain"] = (
    f1_df["grid"] -
    f1_df["positionOrder"]
)
```

```python
f1_df.loc[
    f1_df["valid_grid"] == False,
    "position_gain"
] = pd.NA
```

```python
f1_df["win_flag"] = (
    f1_df["positionOrder"] == 1
).astype(int)
```

```python
f1_df["podium_flag"] = (
    f1_df["positionOrder"] <= 3
).astype(int)
```

```python
f1_df["top10_flag"] = (
    f1_df["positionOrder"] <= 10
).astype(int)
```

```python
f1_df["pole_flag"] = (
    f1_df["qualifying_position"] == 1
).astype(int)
```

```python
f1_df["pole_conversion_flag"] = (
    (f1_df["pole_flag"] == 1) &
    (f1_df["win_flag"] == 1)
).astype(int)
```

```python
f1_df["qualifying_finish_delta"] = (
    f1_df["qualifying_position"] -
    f1_df["positionOrder"]
)
```

```python
f1_df.to_csv(
    "../data/processed/f1_features_dataset.csv",
    index=False
)
```
